### opposition and ground





```python
def get_average_runs_scored_by_team_vs_opposition(df):
    team_opposition_runs_df = df.groupby(['team', 'opposition'])['runs'].mean().reset_index()
    return team_opposition_runs_df
```

```python
def get_average_wickets_taken_by_player_vs_opposition(df):
    player_opposition_wickets_df = df.groupby(['player', 'opposition'])['wkts'].mean().reset_index()
    return player_opposition_wickets_df
```

```python
team_opposition_runs_df = get_average_runs_scored_by_team_vs_opposition(df)
```

```python
sns.heatmap(team_opposition_runs_df.pivot_table(index='team', columns='opposition', values='runs'), annot=True)
plt.title('Average Runs Scored by Team vs Opposition')
plt.xlabel('Opposition')
plt.ylabel('Team')
plt.show()
```

```python
player_opposition_wickets_df = get_average_wickets_taken_by_player_vs_opposition(df)
```

```python
sns.heatmap(player_opposition_wickets_df.pivot_table(index='player', columns='opposition', values='wkts'), annot=True)
plt.title('Average Wickets Taken by Player vs Opposition')
plt.xlabel('Opposition')
plt.ylabel('Player')
plt.show()
```

```python
def get_average_runs_scored_by_team_on_ground(df):
    team_ground_runs_df = df.groupby(['team', 'ground'])['runs'].mean().reset_index()
    return team_ground_runs_df
```

```python
def get_average_wickets_taken_by_player_on_ground(df):
    player_ground_wickets_df = df.groupby(['player', 'ground'])['wkts'].mean().reset_index()
    return player_ground_wickets_df
```

```python
team_ground_runs_df = get_average_runs_scored_by_team_on_ground(df)
```

```python
sns.heatmap(team_ground_runs_df.pivot_table(index='team', columns='ground', values='runs'), annot=True)
plt.title('Average Runs Scored by Team vs Ground')
plt.xlabel('Ground')
plt.ylabel('Team')
plt.show()
```

```python
player_ground_wickets_df = get_average_wickets_taken_by_player_on_ground(df)
```

```python
sns.heatmap(player_ground_wickets_df.pivot_table(index='player', columns='ground', values='wkts'), annot=True)
plt.title('Average Wickets Taken by Player vs Ground')
plt.xlabel('Ground')
plt.ylabel('Player')
plt.show()
```

```python
sns.boxplot(
    x = "opposition",
    y = "runs",
    showmeans=True,
    data=team_opposition_runs_df
)
plt.title("Distribution of Runs Scored by Teams vs Opposition")
plt.xlabel("Opposition")
plt.ylabel("Runs")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
sns.violinplot(
    x = "opposition",
    y = "wkts",
    showmeans=True,
    data=player_opposition_wickets_df
)
plt.title("Distribution of Wickets Taken by Players vs Opposition")
plt.xlabel("Opposition")
plt.ylabel("Wickets")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
sns.scatterplot(
    x = "ground",
    y = "runs",
    data=team_ground_runs_df
)
plt.title("Runs Scored by Teams on Different Grounds")
plt.xlabel("Ground")
plt.ylabel("Runs")
plt.xticks(rotation=45, ha='right')
plt.show()
```
